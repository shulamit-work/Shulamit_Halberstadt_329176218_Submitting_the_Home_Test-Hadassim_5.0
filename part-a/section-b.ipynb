```python
def get_avgs_for_day(df):
    sort_df = df.sort_values (by='timestamp', ignore_index=True)
    just_df = pd.DataFrame(sort_df, columns=['timestamp', 'value'])
    grouped = just_df.groupby(pd.Grouper(key='timestamp', axis=0, freq='H')).mean()
    grouped.reset_index(inplace=True)
    grouped.rename(columns={"timestamp": "start time", "value": "avg value"}, inplace=True)
    grouped.head(5)
    return grouped
print ('function is ready')
```
